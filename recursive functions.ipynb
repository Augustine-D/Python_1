```python
#factorial of a number
```

```python
def factorial(x):
    if x== 1:
        return 1
    else:
        return x*factorial(x-1)
```

```python
factorial(15)
```

```python
factorial(5)
```

```python
# 0 1 1 2 3 5 8...... addistioon
```

```python
def fibinacci(x):
    if x==1:
        return 0
    elif x ==2:
        return 1
    else:
        return (fibinacci(x-1) + fibinacci(x-2))
```

```python
fibinacci(1)
```

```python
fibinacci(2)
```

```python
fibinacci(3)
```

```python
fibinacci(4)
```

```python
fibinacci(5)
```

```python
x = int(input('enter number for fibinacci series'))
l1=[]
for i in range(1,x+1):
    l1.append(fibinacci(i))
l1
```
