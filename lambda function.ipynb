# lambda function

```python
# helps in creating small anonymous functions
#takes any nuber of arguments but one expression
```

```python
def square(x):
    return(x*x)
```

```python
square(2)
```

```python
square = lambda x: x*x
```

```python
square(20)
```

```python
cube = lambda x: x**3
```

```python
cube (3)
```

```python
def addition(x,y):
    add = x+y
    return (add)
```

```python
addition(20,20)
```

```python
add = lambda x,y,z: x+y+z
```

```python
add(1,1,1)
```

```python
def greater(x):
    if x > 15:
        print ('>15')
        
    else:
        print('less than 15')
```

```python
greater(100)
```

```python
greater = lambda x: 'greater than 15' if x>15 else 'lesser'
```

```python
greater(2)
```
