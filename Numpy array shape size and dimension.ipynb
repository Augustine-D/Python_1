```python
arr3 = np.array([1,2,3])
```

```python
print(arr3.ndim)
print(arr3.size)
print(arr3.shape)
```

```python
arr4 = np.array([[1,2],[1,2]])
```

```python
print(arr4.ndim)
print(arr4.size)
print(arr4.shape)
```

```python
arr5 = np.array([[1,2,3],[1,2,3]])
```

```python
print(arr5.ndim)
print(arr5.size)
print(arr5.shape)
```

```python
arr6 = np.array([[[1,2],[1,2],[1,2]],[[1,2],[1,2],[1,2]],[[1,2],[1,2],[1,2]]])
```

```python
print(arr6.ndim)
print(arr6.size)
print(arr6.shape)
```
