# map function

```python
#map function takes a functions and iter(list/tuple) as an argument and returns a map object of all the results, which is an 
#iterator after applying the function to all the elements in iter
```

```python
x = [1,2,3,4,5,6]
```

```python
def square(x):
    return (x**2)
```

```python
y = map(square,x)
```

```python
y
```

```python
list(y)
```

```python
def add(x):
    return(x+100)
```

```python
x = [1,2,3,4,5,6]
y = [10,20,30,40,50]
```
